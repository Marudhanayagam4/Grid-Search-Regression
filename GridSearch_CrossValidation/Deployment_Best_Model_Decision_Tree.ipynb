```python
import pickle
scx=pickle.load(open("Standard_Scaler_X.pkl",'rb'))
scy=pickle.load(open("Standard_Scaler_Y.pkl", 'rb'))
scy
```

```python
filename="final_model.sav"
model = pickle.load(open(filename, 'rb'))
model
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))

preInput=scx.transform([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
preInput
```

```python
result=model.predict(preInput)
result
```

```python
preOutput=scy.inverse_transform([result])
print("Future Prediction = {}".format(preOutput))
```
